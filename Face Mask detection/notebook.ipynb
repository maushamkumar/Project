```python
import os
from keras.preprocessing import image
```

```python
import cv2
```

```python
categories = ['with_mask', 'without_mask']
```

```python
data = []

for category in categories:
    path = os.path.join('data', category)
    label = categories.index(category)
    
    for file in os.listdir(path):
        img_path = os.path.join(path, file)
        img = cv2.imread(img_path)
        img = cv2.resize(img, (224, 224))
        # Images are numpy arrary 
        # At the end we have create a data that have two coloumn 
        # one column have numpy array and second column have label 
        data.append([img, label])
        
```

```python
import os

data_path = "data"
found = {"with_mask": False, "without_mask": False}

for folder in os.listdir(data_path):
    folder_path = os.path.join(data_path, folder)
    
    if os.path.isdir(folder_path) and folder in found:
        for file in os.listdir(folder_path):
            if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                print(f"{folder} Image: {os.path.join(folder_path, file)}")
                found[folder] = True
                break

    # Break early if both found
    if all(found.values()):
        break
```

```python
len(data)
```

```python
# Now we will shuffle the data 
import random
```

```python
random.shuffle(data)
```

```python
X = []
y = []

for feature, label in data: 
    X.append(feature)
    y.append(label)
```

```python
len(X)
```

```python
len(y)
```

```python
import numpy as np 
```

```python
X = np.array(X)
y = np.array(y)
```

```python
X.shape
```

```python
# Scaling 
X = X/255
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from keras.applications.vgg16 import VGG16
```

```python
vgg = VGG16()
```

```python
vgg.summary()
```

```python
from keras import Sequential
```

```python
model = Sequential()
```

```python
for layer in vgg.layers[:-1]:
    model.add(layer)
```

```python
model.summary()
```

```python
for layer in model.layers:
    layer.trainable = False
```

```python
model.summary()
```

```python
from keras.layers import Dense
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=3, validation_data=(X_test, y_test))
```
